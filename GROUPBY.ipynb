## Q1:how many free apps are there in ART_AND_DESIGN

```python
a=len(df[(df["Category"] == "ART_AND_DESIGN") & (df["Type"] == "Free")])
print("There are ", a , " free apps in ART_AND_DESIGN category")
```

## Q2: How many apps are there in ART_AND_DESIGN with rating more than 4.5

```python
b=len(df[(df["Category"] == "ART_AND_DESIGN") & (df["Rating"] > 4.5)])
print("There are ", b , " apps in ART_AND_DESIGN category with rating more than 4.5")
```

## Q3 How many apps are there in FAMILY with rating more than 4.5 and are free

```python
 c = len(df[(df["Category"] == "FAMILY") & (df["Rating"] > 4.5) & (df["Type"] == "Free")])
print("There are ", c , "  free apps in FAMILY category with rating more than 4.5")
```

## Q4 : List all free apps with rating more than 4.5 in family category

## Q5: Print the names of the apps in ART_AND_DESIGN that are having rating more than 4.5 and sort them WRT to rating in ascending orders

```python
df_pr=df[df["Category"] == "ART_AND_DESIGN"]
df_pr = df_pr[df_pr["Rating"] > 4.5]
df_pr.sort_values(by  = "Rating" , ascending = True).reset_index().head(10)
```

## Q6: Print the names of the apps in ART_AND_DESIGN that are having rating more than 4.5 and sort them WRT to rating in descending orders

```python
df_pr=df[df["Category"] == "ART_AND_DESIGN"]
df_pr = df_pr[df_pr["Rating"] > 4.5]
df_new = df_pr.sort_values(by  = "Rating" , ascending = False).reset_index()
df_new
```
